```python
import numpy as np
import cv2 as cv 
import matplotlib.pyplot as plt
```

```python
img = cv.imread("../DATA/image1.png")
cv.imshow("Image" , img )
cv.waitKey(0)
```

```python
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow("Gray" , gray_img )
cv.waitKey(0)
```

```python
## Create Prewitt Filter

kernelx = np.array([[-1, 0 , 1 ], [-1, 0 , 1]  , [-1 , 0 , 1 ]])
kernely = np.array([[1,1,1] , [ 0,0,0] , [-1,-1,-2]])

img_prewittx = cv.filter2D(img, -1, kernelx)
img_prewitty = cv.filter2D(img, -1, kernely)

prewitt = np.hypot(img_prewitty, img_prewittx)
prewitt = prewitt[:,:,0]
prewitt = prewitt.astype("int")
plt.imshow(prewitt , cmap='gray')
```

```python
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
kernelx = np.array([
    [-1,0,1],
    [-2,0,2],
    [-1,0,1]
])
kernely = np.array([
    [1,2,1],
    [0,0,0],
    [-1,-2,-1]
])
img_x = cv.filter2D(gray_img, -1, kernelx)
img_y = cv.filter2D(gray_img, -1, kernely)
new=np.hypot(img_x,img_y)
plt.imshow(new.astype('int'),cmap='gray')
```

```python
image = img

(H,W) = image.shape[:2]

gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
blurred = cv.GaussianBlur(gray , (5,5) , 0)
canny = cv.Canny(blurred, 50,250)
fig, ax = plt.subplots(1,2,figsize=(18,18))
ax[0].imshow(gray, cmap='gray')
ax[1].imshow(canny, cmap='gray')

ax[0].axis('off')
ax[1].axis('off')
```
