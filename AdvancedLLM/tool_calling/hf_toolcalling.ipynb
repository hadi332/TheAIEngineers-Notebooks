```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

checkpoint =  "mistralai/Mistral-7B-Instruct-v0.3"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = AutoModelForCausalLM.from_pretrained(checkpoint, torch_dtype=torch.float16, device_map="auto")
```

```python
def get_current_tempreate(location :str, unit: str) -> float:
  """
    Get the current temperature at a location.

    Args:
        location: The location to get the temperature for, in the format "City, Country"
        unit: The unit to return the temperature in. (choices: ["celsius", "fahrenheit"])
    Returns:
        The current temperature at the specified location in the specified units, as a float.
    """
  return 22
```

```python
def get_current_wind_speed(location: str) -> float:
    """
    Get the current wind speed in km/h at a given location.

    Args:
        location: The location to get the temperature for, in the format "City, Country"
    Returns:
        The current wind speed at the given location in km/h, as a float.
    """
    return 6.
```

```python
import torch
import random
import string
from transformers import AutoModelForCausalLM, AutoTokenizer


# Load model and tokenizer
model_id = "mistralai/Mistral-7B-Instruct-v0.3"
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id, torch_dtype=torch.bfloat16, device_map="auto"
)


# Define the tool
def get_current_weather(location: str):
    """
    Get the current weather

    Args:
        location: The city and state, e.g. San Francisco, CA
    """
    return f"The weather in {location} is 22°C with clear skies."  # Mocked response


tools = [get_current_weather]


# Helper to create random tool call IDs
def generate_tool_call_id():
    return ''.join(random.choices(string.ascii_lowercase + string.digits, k=9))


# Chat loop
def chat():
    conversation = [{"role": "system", "content": "You are a weather bot. Use tools when needed."}]

    while True:
        user_message = input("\n🧑 You: ")
        if user_message.lower() in ["exit", "quit"]:
            print("👋 Goodbye!")
            break

        conversation.append({"role": "user", "content": user_message})

        # First generation (model may request a tool)
        inputs = tokenizer.apply_chat_template(
            conversation,
            tools=tools,
            add_generation_prompt=True,
            return_tensors="pt"
        ).to(model.device)
        outputs = model.generate(input_ids=inputs, max_new_tokens=512)
        raw_response = tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:], skip_special_tokens=True)

        print(f"\n🤖 Model output:\n{raw_response}")

        # Simulate tool call (in a real scenario, you'd parse this from the model output)
        #TODO: Do your research and write it in a real scenario
        tool_call_id = generate_tool_call_id()
        tool_name = "get_current_weather"
        tool_args = {"location": user_message, "format": "celsius"}  # Example args

        print(f"\n🛠️ Tool called: {tool_name} with args {tool_args}")

        # Add tool call to conversation
        tool_call = {
            "name": tool_name,
            "arguments": tool_args
        }
        conversation.append({
            "role": "assistant",
            "tool_calls": [{
                "id": tool_call_id,
                "type": "function",
                "function": tool_call
            }]
        })

        # Execute tool
        tool_result = get_current_weather(**tool_args)
        print(f"\n✅ Tool result: {tool_result}")

        # Add tool result to conversation
        conversation.append({
            "role": "tool",
            "tool_call_id": tool_call_id,
            "name": tool_name,
            "content": tool_result
        })

        # Second generation with tool result
        inputs = tokenizer.apply_chat_template(
            conversation,
            tools=tools,
            add_generation_prompt=True,
            return_tensors="pt"
        ).to(model.device)

        outputs = model.generate(
            input_ids=inputs,
            max_new_tokens=512
        )

        outputs = model.generate(**inputs, max_new_tokens=512)
        final_response = tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:], skip_special_tokens=True)

        print(f"\n🤖 Final response:\n{final_response}")

        # Add final assistant message to the conversation
        conversation.append({"role": "assistant", "content": final_response})


if __name__ == "__main__":
    chat()
```
